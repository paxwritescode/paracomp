```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Solution plot

```python
try:
    data = pd.read_csv('../code/src/results/solution.csv', header=None).values
except FileNotFoundError:
    print("Error: no file")
    exit()
```

```python
Nx = data.shape[0] - 1
Ny = data.shape[1] - 1

x = np.linspace(0, np.pi, Nx + 1)
y = np.linspace(0, 1.0, Ny + 1)
X, Y = np.meshgrid(x, y, indexing='ij')

fig = plt.figure(figsize=(14, 6))
plt.rcParams.update({
    'font.size': 11,
    'font.family': 'serif',
    'text.usetex': False
})


ax1 = fig.add_subplot(1, 2, 1)
contour = ax1.contourf(X, Y, data, levels=50, cmap='coolwarm')
cbar1 = fig.colorbar(contour, ax=ax1)
cbar1.set_label('$u(x, y)$', rotation=0, labelpad=15)
ax1.set_title('2D Numerical Solution (Contour Plot)', fontsize=13, fontweight='bold')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.grid(True, linestyle=':', alpha=0.5)


ax2 = fig.add_subplot(1, 2, 2, projection='3d')
surf = ax2.plot_surface(X, Y, data, cmap='coolwarm', edgecolor='none', alpha=0.9)
cbar2 = fig.colorbar(surf, ax=ax2, shrink=0.6, aspect=10)
cbar2.set_label('$u(x, y)$', rotation=0, labelpad=15)

ax2.set_title('3D Numerical Solution Surface', fontsize=13, fontweight='bold')
ax2.set_xlabel('$x$')
ax2.set_ylabel('$y$')
ax2.set_zlabel('$u(x, y)$')


ax2.view_init(elev=25, azim=135)

plt.tight_layout()

plt.savefig('solution_plots.pdf', dpi=300)
plt.show()
```

Parallel implementation study

```python
try:
    df = pd.read_csv('../code/src/results/scaling_data.csv')
except FileNotFoundError:
    print("Error: no file")
    exit()
```

```python
plt.rcParams.update({
    'font.size': 11,
    'axes.labelsize': 12,
    'axes.titlesize': 14,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'figure.titlesize': 16
})

num_procs = df['NumProcs']

# time on process count
plt.figure(figsize=(7, 5))
plt.plot(num_procs, df['TimeSeconds'], marker='o', color='#1f77b4', linewidth=2)
plt.title('Execution time vs. Process Count')
plt.xlabel('$p$')
plt.ylabel('$T$, s')
plt.xticks(num_procs)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.savefig('time_plot.pdf', dpi=300)
plt.show()
plt.close()

# speedup on process count
plt.figure(figsize=(7, 5))
plt.plot(num_procs, df['Speedup'], marker='o', color='#ff7f0e', linewidth=2)

plt.title('Speedup vs. Process Count')
plt.xlabel('$p$')
plt.ylabel('$S_p$')
plt.xticks(num_procs)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.savefig('speedup_plot.pdf', dpi=300)
plt.show()
plt.close()

# efficiency on process count
plt.figure(figsize=(7, 5))
plt.plot(num_procs, df['Efficiency'], marker='o', color='#2ca02c', linewidth=2)

plt.title('Efficiency vs. process count')
plt.xlabel('$p$')
plt.ylabel('$E_p$')
plt.xticks(num_procs)
plt.ylim(0, 1.1)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.savefig('efficiency_plot.pdf', dpi=300)
plt.show()
plt.close()
```

```python
columns = ['Nx', 'Ny', 'NumProcs', 'TimeSeconds']
try:
    df = pd.read_csv('../code/src/results/time_vs_grid.csv', names=columns)
except FileNotFoundError:
    print("Error: no file")
    exit()
```

```python
df = df.groupby(['Nx', 'Ny', 'NumProcs'], as_index=False)['TimeSeconds'].mean()

plt.rcParams.update({
    'font.size': 11,
    'axes.labelsize': 12,
    'axes.titlesize': 13,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'grid.linestyle': '--',
    'grid.alpha': 0.6
})
```

```python
# plot 1: Nx fixed
plt.figure(figsize=(7, 5))

df_nx40 = df[df['Nx'] == 40].sort_values('Ny')
plt.plot(df_nx40['Ny'], df_nx40['TimeSeconds'], marker='o', color='#1f77b4', linewidth=2, label='$N_x = 40$')

df_nx100 = df[df['Nx'] == 100].sort_values('Ny')
plt.plot(df_nx100['Ny'], df_nx100['TimeSeconds'], marker='s', color='#d62728', linewidth=2, label='$N_x = 100$')

plt.title('Execution Time vs. Grid Size $N_y$')
plt.xlabel('$N_y$')
plt.ylabel('T, s')
plt.xlim(0, 210)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('time_vs_ny.pdf', dpi=300)
plt.show()
plt.close()

# plot 2: Ny fixed
plt.figure(figsize=(7, 5))

df_ny40 = df[df['Ny'] == 40].sort_values('Nx')
plt.plot(df_ny40['Nx'], df_ny40['TimeSeconds'], marker='o', color='#2ca02c', linewidth=2, label='$N_y = 40$')

df_ny100 = df[df['Ny'] == 100].sort_values('Nx')
plt.plot(df_ny100['Nx'], df_ny100['TimeSeconds'], marker='s', color='#9467bd', linewidth=2, label='$N_y = 100$')

plt.title('Execution Time vs. Grid Size $N_x$')
plt.xlabel('$N_x$')
plt.ylabel('T, s')
plt.xlim(0, 210)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('time_vs_nx.pdf', dpi=300)
plt.show()
plt.close()
```
