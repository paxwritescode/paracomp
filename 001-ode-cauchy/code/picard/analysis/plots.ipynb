```python
import csv
import matplotlib.pyplot as plt
```

Time vs size

```python
with open("../results/time_vs_size.csv", encoding="utf-8") as f_time_vs_size:
    for line in f_time_vs_size:
        if line.strip().startswith("#") or not line.strip():
            continue
        time_vs_size_list = [
            float(x.strip())
            for x in line.split(",")
            if x.strip()
        ]

with open("../results/time_vs_threads.csv", encoding="utf-8") as f_time_vs_threads:
    for line in f_time_vs_threads:
        if line.strip().startswith("#") or not line.strip():
            continue
        time_vs_threads_list = [
            float(x.strip())
            for x in line.split(",")
            if x.strip()
        ]

with open("../results/speedup_vs_threads.csv", encoding="utf-8") as f_speedup_vs_threads:
    for line in f_speedup_vs_threads:
        if line.strip().startswith("#") or not line.strip():
            continue
        speedup_vs_threads_list = [
            float(x.strip())
            for x in line.split(",")
            if x.strip()
        ]

with open("../results/efficiency_vs_threads.csv", encoding="utf-8") as f_efficiency_vs_threads:
    for line in f_efficiency_vs_threads:
        if line.strip().startswith("#") or not line.strip():
            continue
        efficiency_vs_threads_list = [
            float(x.strip())
            for x in line.split(",")
            if x.strip()
        ]

size_list = [10 * k for k in range(1, 20 + 1)]
threads_list = [k for k in range(1, 12 + 1)]

plt.plot(size_list, time_vs_size_list)
plt.xlabel("n")
plt.ylabel("t")
plt.grid()
plt.title("Execution time on SLAE dimensionality")
plt.savefig("../results/time_vs_size.pdf")
plt.show()

plt.plot(threads_list, time_vs_threads_list)
plt.xlabel("threads")
plt.ylabel("t")
plt.grid()
plt.title("Execution time on threads number")
plt.savefig("../results/time_vs_threads.pdf")
plt.show()

plt.plot(threads_list, speedup_vs_threads_list)
plt.xlabel("threads")
plt.ylabel("speedup")
plt.grid()
plt.title("Speedup on threads number")
plt.savefig("../results/speedup_vs_threads.pdf")
plt.show()

plt.plot(threads_list, efficiency_vs_threads_list)
plt.xlabel("threads")
plt.ylabel("parallelization efficiency")
plt.grid()
plt.title("Parallelization efficiency on threads number")
plt.savefig("../results/efficiency_vs_threads.pdf")
plt.show()
```
