```python
import csv
import matplotlib.pyplot as plt
```

```python
def read_csv_line_as_floats(path):
    with open(path, encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith("#"):
                continue
            return [float(x) for x in line.split(",") if x.strip()]
```

```python
time_vs_size_list = read_csv_line_as_floats("../results/time_vs_size.csv")
time_vs_threads_list = read_csv_line_as_floats("../results/time_vs_threads.csv")
speedup_vs_threads_list = read_csv_line_as_floats("../results/speedup_vs_threads.csv")
efficiency_vs_threads_list = read_csv_line_as_floats("../results/efficiency_vs_threads.csv")
```

```python
def plot_and_save(x, y, xlabel, ylabel, title, filename):
    plt.plot(x, y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid()
    plt.savefig(filename)
    plt.show()
```

```python
size_list = [10 * k for k in range(1, 31)]
threads_list = list(range(1, 13))
```

```python
plot_and_save(
    size_list,
    time_vs_size_list,
    xlabel="n",
    ylabel="t",
    title="Execution time on SLAE dimensionality",
    filename="../results/time_vs_size.pdf"
)
```

```python
plot_and_save(
    threads_list,
    time_vs_threads_list,
    xlabel="threads",
    ylabel="t",
    title="Execution time on threads number",
    filename="../results/time_vs_threads.pdf"
)

plot_and_save(
    threads_list,
    speedup_vs_threads_list,
    xlabel="threads",
    ylabel="speedup",
    title="Speedup on threads number",
    filename="../results/speedup_vs_threads.pdf"
)

plot_and_save(
    threads_list,
    efficiency_vs_threads_list,
    xlabel="threads",
    ylabel="parallelization efficiency",
    title="Parallelization efficiency on threads number",
    filename="../results/efficiency_vs_threads.pdf"
)
```
